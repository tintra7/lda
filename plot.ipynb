```python
import matplotlib.pyplot as plt
import numpy as np

# Data
datasets = ['82k', '194k']
sklearn_time = [24.73, 62.17]
custome_time = [7.54, 16.77]
sklearn_mem = [2338.77 , 5603.12]
custome_mem = [286.71, 667.55]

# Plot Configuration
x = np.arange(len(datasets))  # Label locations
width = 0.35  # Width of bars

# Create subplots for time and memory
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Time Plot
ax[0].bar(x - width/2, sklearn_time, width, label='Sklearn LDA', color='blue', alpha=0.7)
ax[0].bar(x + width/2, custome_time, width, label='Custom LDA', color='orange', alpha=0.7)
ax[0].set_title('Execution Time Comparison')
ax[0].set_ylabel('Time (seconds)')
ax[0].set_xticks(x)
ax[0].set_xticklabels(datasets)
ax[0].legend()
ax[0].grid(axis='y', linestyle='--', alpha=0.6)

# Memory Plot
ax[1].bar(x - width/2, sklearn_mem, width, label='Sklearn LDA', color='blue', alpha=0.7)
ax[1].bar(x + width/2, custome_mem, width, label='Custom LDA', color='orange', alpha=0.7)
ax[1].set_title('Memory Usage Comparison')
ax[1].set_ylabel('Memory (MiB)')
ax[1].set_xticks(x)
ax[1].set_xticklabels(datasets)
ax[1].legend()
ax[1].grid(axis='y', linestyle='--', alpha=0.6)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
